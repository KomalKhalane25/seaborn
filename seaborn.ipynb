```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
tips_df = sns.load_dataset('tips')
```

```python
#countplot()

sns.countplot(x='day', data=tips_df)
```

```python
sns.countplot(x='sex',data=tips_df)
```

```python
tips_df['sex'].value_counts()
```

```python
tips_df.groupby('sex').count()
```

```python
sns.countplot(x='smoker', data=tips_df)
```

```python
sns.countplot(x='sex', data=tips_df,hue='smoker')
```

```python
tips_df.groupby('sex')['smoker'].value_counts()
```

```python
tips_df.groupby(['sex','smoker']).count()
```

```python
tips_df[['sex','smoker']].value_counts()
```

```python
sns.countplot(x='sex', data=tips_df,hue='smoker',palette='coolwarm')
```

```python
sns.countplot(x='sex', data=tips_df,hue='smoker',palette='rainbow')
```

```python
#barplot
sns.barplot(x='sex', y='total_bill', data=tips_df)
```

```python
tips_df.groupby('sex').mean('total_bill')
```

```python
tips_df[['sex','total_bill']].value_counts()
```

```python
sns.barplot(x='time', y='total_bill', data=tips_df)
```

```python
sns.barplot(x='time', y='total_bill', data=tips_df,order=['Dinner','Lunch'])
```

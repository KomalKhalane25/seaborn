```python
sns.countplot(x='sex', data=tips_df,hue='smoker',palette='rainbow')
```

```python
#barplot
sns.barplot(x='sex', y='total_bill', data=tips_df)
```

```python
tips_df.groupby('sex').mean('total_bill')
```

```python
tips_df[['sex','total_bill']].value_counts()
```

```python
sns.barplot(x='time', y='total_bill', data=tips_df)
```

```python
sns.barplot(x='time', y='total_bill', data=tips_df,order=['Dinner','Lunch'])
```

```python
tips_df['weekend'] = tips_df['day'].isin(['Sat','Sun'])
```

```python
tips_df.tail(20)
```

```python
sns.barplot(x='day', y='total_bill', data=tips_df,hue='weekend')
```

```python
#stripplot
sns.stripplot(x='day', y='total_bill', data=tips_df)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips_df ,hue='sex')
```

```python
sns.stripplot(x='day', y='total_bill', data=tips_df,hue='sex',dodge=True)
```

```python
#swarmpot()
sns.swarmplot(x='day', y='total_bill', data=tips_df,hue='sex')
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips_df,hue='sex',dodge=True,palette='Set2')
```

```python
#heatmap()
flight_df = sns.load_dataset('flights')
```

```python
flight_df.info()
```

```python
flight_df.head(20)
```

```python
flight_df.tail(20)
```

```python
pvflight = flight_df.pivot_table(index='month',columns='year',values='passengers')
pvflight
```

```python
sns.heatmap(pvflight,linecolor='white',linewidths=1)
```

```python
sns.heatmap(pvflight,linecolor='white',linewidths=2,cmap='crest')
```

```python
#grids
#facetgrid
#pairgrid
```

```python
iris_df = sns.load_dataset('iris')
```

```python
iris_df.info()
```

```python
#PairGrid
import matplotlib.pyplot as plt
sns.PairGrid(iris_df).map(plt.scatter)
```

```python
g = sns.PairGrid(iris_df)
g.map_diag(plt.hist)
g.map_upper(plt.bar)
g.map_lower(plt.scatter)
```

```python
#facetGrid
sns.FacetGrid(tips_df,col='day',row='smoker').map(plt.scatter,'total_bill','tip')
```

```python
sns.FacetGrid(tips_df,col='smoker',row='day').map(plt.scatter,'total_bill','tip')
```
